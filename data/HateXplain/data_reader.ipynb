```python
import pandas as pd
import json
from tqdm import tqdm
```

```python
# Read the data
with open('data.json') as f:
    data = json.load(f)
```

```python
data = pd.DataFrame(data).T
```

```python
data
```

```python
for index, row in tqdm(data.iterrows(), total=data.shape[0]):
    # get the majority label
    labels = []
    targets = []
    for d in row['annotators']:
        labels.append(d['label'])
        targets += d['target']
    # find the most common label
    label = max(set(labels), key=labels.count)
    # find the most common target
    target = max(set(targets), key=targets.count)
    # save the label and target to new columns
    data.loc[index, 'label'] = label
    data.loc[index, 'target'] = target
    # print(label, target)

    # transform post tokens to string
    # replace , with space and remove the first and last character
    data.loc[index, 'text'] = ' '.join(row['post_tokens']).replace(',','')
```

```python
# reset the index
data.reset_index(drop=True, inplace=True)
```

```python
# drop annotators, rationales and post_tokens columns
data.drop(['annotators', 'rationales', 'post_tokens'], axis=1, inplace=True)
```

```python
# for each sentence, we equally seperate the sentence into two parts
def split_sentence(sentence):
    tokens = sentence.split()
    split_point = len(tokens) // 2
    prefix = ' '.join(tokens[:split_point])
    suffix = ' '.join(tokens[split_point:])
    return prefix, suffix
```

```python
# remove the records if the prefix is shorter than 3 words
data['prefix'], data['suffix'] = zip(*data['text'].apply(split_sentence))
data = data[data['prefix'].apply(lambda x: len(x.split()) > 3)]
data.reset_index(drop=True, inplace=True)
```

```python
# split the data according to the label
data_0 = data[data['label'] == 'normal']
data_1 = data[data['label'] == 'hatespeech']
data_2 = data[data['label'] == 'offensive']
```

```python
# combine the data1 and data2, only keep id, prefix, label and target
data_1_2 = pd.concat([data_1, data_2], axis=0)
data_1_2 = data_1_2[['post_id', 'prefix', 'label', 'target']]
data_1_2.reset_index(drop=True, inplace=True)
```
